###**3. Cấu hình model YOLOv4**

- darknet/cfg -> tìm và mở file yolo4-custom.cfg và chỉnh sửa.
- Dòng 3: Batch = 64
- Dòng 4: subdivisions=32
- Tìm đến dòng 20, sửa max_batches = max(<số class>*2000,6000).
- Đến dòng 22 sửa thành steps=80%, 90% của max_batches. Ví dụ ở đây là steps=4800,5400.
- Replace toàn bộ các dòng có “classes=80” thành “classes=<số class>”. Nó ở các dòng 970, 1058 và 1146.
- Replace toàn bộ các dòng có “filters=255” thành “filters=<(số class+5)*3>”. Nó ở các dòng: 963, 1051 và 1139.
- Nếu bị out memory, hãy chuyển subdivisions (dòng 7) thành 32 (hoặc 64) hoặc các bạn giảm size ảnh xuống còn width=416, height=416 (dòng 8,9).
